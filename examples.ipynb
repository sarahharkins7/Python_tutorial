```python
import pandas as pd
elections = pd.read_csv('Daily_Kos_Elections_08_12_16_congress_districts.csv')
elections
```

```python
# Chaging the indexing from the leftmost column 0, 1, 2, ... to the congressional district 
elections = elections.set_index(elections['CD'])
```

```python
# To access specific rows, use 'loc' and 'iloc'
elections.loc['AK-AL']
```

```python
elections.loc['AL-01', ['Incumbent', 'Party']]
```

```python
elections.iloc[2,:]
```

```python
# Boolean Indxing in a Pandas DataFrame
elections[elections['Trump\n2016'] < elections['Romney\n2012']]
```

```python
elections[np.logical_and(elections['Trump\n2016'] > 50, elections['Obama\n2012'] >50)]
```

```python
numeric_cols = elections.select_dtypes(include='number')
numeric_cols
```

```python
filtered_df = elections.loc[np.logical_and(elections['Trump\n2016'] > 50, elections['Obama\n2012'] > 50), 'Incumbent']
filtered_df
```

```python
df = elections[elections['Party'] == "(D)"]
df.to_excel("democratic_incumbent.xlsx")
```

```python
tennessee = elections[elections['CD'].str.contains('TN')]
tennessee.select_dtypes(include='number')
```

```python
tennessee.iloc[:, 3::2] #starating at column index 3 and skippng evey other columns 
```

```python
tennessee.iloc[:, 3::2].plot(kind = 'bar')
```
